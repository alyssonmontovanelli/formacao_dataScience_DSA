# Com Broadcasting

Propagação de uma operação ao longo do dataframe


```python
# Adicionando o valor "5" a todo o dataframe
df['base_salary'] = df['base_salary'].add(5)
```

```python
# Adicionando nova coluna com Broadcasting (convertendo de USD para EURO, nete exemplo)
df['base_salary_EUR'] = df['base_salary'].mul(0.92)
```

```python
df.head()
```

```python
# Drop de coluna
df.drop(columns = ['base_salary_EUR'], inplace = True)
```

```python
df.head()
```

```python
# Adicionando coluna de EURO ao lado de base_salary
df.insert(6, column = "base_salary_EUR", value = df['base_salary'].mul(0.92))
```

```python
df.head()
```

```python
# Rename 
df.rename(columns = {'base_salary': 'base_salary_USD',
                    'guaranteed_compensation': 'guaranteed_compensation_USD',
                    'salario_final': 'salario_final_USD'},
         inplace = True)
```

```python
df.head()
```

```python
# Convertendo a coluna ID em índex da tabela

df.set_index('ID', inplace = True)
```
